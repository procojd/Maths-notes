# üìò Jacobians ‚Äì Notes

### ‚ú® Definition
If we have functions:

$$u = u(x, y), \quad v = v(x, y)$$

Then the **Jacobian** of $(u,v)$ with respect to $(x,y)$ is defined as:

$$ J = \frac{\partial(u,v)}{\partial(x,y)} = \begin{vmatrix}
\frac{\partial u}{\partial x} & \frac{\partial u}{\partial y} \\
\frac{\partial v}{\partial x} & \frac{\partial v}{\partial y}
\end{vmatrix} $$

This determinant is called the **Jacobian determinant**.

---
### üîπ Example 1
Let

$$ u = x^2 + y^2, \quad v = x^2 - y^2 $$

Then,

$$ \frac{\partial u}{\partial x} = 2x, \quad \frac{\partial u}{\partial y} = 2y $$
$$ \frac{\partial v}{\partial x} = 2x, \quad \frac{\partial v}{\partial y} = -2y $$

So,

$$ J = \begin{vmatrix} 2x & 2y \\ 2x & -2y \end{vmatrix} = (2x)(-2y) - (2y)(2x) = -8xy $$

---
### üîπ Properties of Jacobians
1. **Chain Rule:**

   If $u = u(x,y), v = v(x,y)$ and $x = x(r,s), y = y(r,s)$, then

   $$ \frac{\partial(u,v)}{\partial(r,s)} = \frac{\partial(u,v)}{\partial(x,y)} \cdot \frac{\partial(x,y)}{\partial(r,s)} $$

2. If $u = x, v = y$, then $J = 1$.

3. If $u, v$ are interchanged, then $J$ changes sign.

---
### üîπ Example 2 ‚Äì Chain Rule Application
Let

$$ u = x+y, \quad v = x-y $$
and
$$ x = r \cos s, \quad y = r \sin s $$

Step 1: Find $\frac{\partial(u,v)}{\partial(x,y)}$

$$ \begin{vmatrix}
\frac{\partial u}{\partial x} & \frac{\partial u}{\partial y} \\
\frac{\partial v}{\partial x} & \frac{\partial v}{\partial y}
\end{vmatrix} = \begin{vmatrix} 1 & 1 \\ 1 & -1 \end{vmatrix} = -2 $$

Step 2: Find $\frac{\partial(x,y)}{\partial(r,s)}$

$$ \begin{vmatrix}
\frac{\partial x}{\partial r} & \frac{\partial x}{\partial s} \\
\frac{\partial y}{\partial r} & \frac{\partial y}{\partial s}
\end{vmatrix} = \begin{vmatrix} \cos s & -r\sin s \\ \sin s & r \cos s \end{vmatrix} $$

$$ = r(\cos^2 s + \sin^2 s) = r $$

Step 3: Apply Chain Rule

$$ \frac{\partial(u,v)}{\partial(r,s)} = \frac{\partial(u,v)}{\partial(x,y)} \cdot \frac{\partial(x,y)}{\partial(r,s)} $$

$$ = (-2)(r) = -2r $$

---
### ‚úÖ Uses of Jacobians
1. Transformation of variables in multiple integrals.
2. Checking functional independence.
3. Useful in coordinate transformations (Cartesian ‚Üí Polar, etc.).
