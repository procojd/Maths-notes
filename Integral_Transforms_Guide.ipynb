## 1Ô∏è‚É£ Laplace Transform

### Definition
For a function $f(t)$ defined for $t \geq 0$,

$$
L\{f(t)\} = F(s) = \int_{0}^{\infty} e^{-st} f(t) \, dt, \quad s \in \mathbb{C}
$$

### Example

$$
L\{e^{at}\} = \int_{0}^{\infty} e^{-st} e^{at} dt = \frac{1}{s-a}, \quad \text{for Re}(s) > a
$$

---

## 2Ô∏è‚É£ Inverse Laplace Transform

### Definition
The inverse Laplace recovers $f(t)$ from $F(s)$:

$$
f(t) = L^{-1}\{F(s)\}
$$

### Example

$$
F(s) = \frac{1}{s+2} \quad \Rightarrow \quad f(t) = e^{-2t}, \; t \geq 0
$$

## 4Ô∏è‚É£ Fourier Analysis

### 4.1 Fourier Series
Expresses a **periodic function** $f(x)$ with period $2\pi$ as:

$$
f(x) = a_0 + \sum_{n=1}^{\infty} \left( a_n \cos(nx) + b_n \sin(nx) \right)
$$

Where:

$$
a_n = \frac{1}{\pi} \int_{-\pi}^{\pi} f(x)\cos(nx) \, dx, \quad
b_n = \frac{1}{\pi} \int_{-\pi}^{\pi} f(x)\sin(nx) \, dx
$$

**Example:**  
Square wave expansion using sine terms only.

---

### 4.2 Fourier Integral
For **non-periodic functions**, Fourier integral generalizes series:

$$
f(x) = \int_{-\infty}^{\infty} F(\omega) e^{i \omega x} d\omega
$$

---

### 4.3 Fourier Transform
Defined as:

$$
F(\omega) = \int_{-\infty}^{\infty} f(t) e^{-i \omega t} dt
$$

Inverse Fourier Transform:

$$
f(t) = \frac{1}{2\pi} \int_{-\infty}^{\infty} F(\omega) e^{i \omega t} d\omega
$$

**Example:**  

$$
f(t) = e^{-at}, \; a>0 \quad \Rightarrow \quad F(\omega) = \frac{1}{a+i\omega}
$$

---

### 4.4 Finite Fourier Transform
When the function is defined only on a **finite interval** $[0, L]$,  
we expand in terms of sine and cosine series restricted to that interval.
